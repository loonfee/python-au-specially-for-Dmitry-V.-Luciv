```python
import pandas
import matplotlib.pyplot as plt
from scipy import stats

df1 = pandas.read_csv("experiments.csv")
#Проверим наличие мировых заговоров на основе количества запросов "брус на наклоннной поверхности"

df1['experiments'].plot(kind='bar')

df1 = pandas.DataFrame(data={
    'df1': df1['experiments'],
})

df1.plot.kde()

d1 = df1['df1']

print(stats.kstest(d1, 'norm', (d1.mean(), d1.std()), N=5000))
#Распределение нормально - можем спать спокойно, или..
```
